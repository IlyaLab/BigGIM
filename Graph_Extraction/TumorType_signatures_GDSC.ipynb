```python
TCGA_Mondo_map = pd.read_csv(input_dir + "Mondo_TCGA.csv")
TCGA_Mondo_map
dic_mondo_map = {}
for i in range(0, TCGA_Mondo_map.shape[0]):
    dic_mondo_map['TCGA_' + list(TCGA_Mondo_map['Study Abbreviation'])[i]] = list(TCGA_Mondo_map['MONDO'])[i]
    dic_mondo_map['TCGA-' + list(TCGA_Mondo_map['Study Abbreviation'])[i]] = list(TCGA_Mondo_map['MONDO'])[i]
```

```python
#Generate an ID mapping dictionary for gene symbol and NCBI gene ID
ID_map = pd.read_csv("/Users/guangrong/Documents/GitHub_project/fastqpi_BigGIM/KGs/ID_convert/Table_ID_convert_9.19.2021.tsv", sep = '\t')
dic_ID_symbol2ncbi ={}
import math
ID_map.astype(str)
for i in range(0,ID_map.shape[0]):
    symbol = ID_map['Approved_symbol'][i]
    
    ncbi_id = ID_map['NCBI_Gene_ID'][i]
    ensemble_id = ID_map['Ensembl_gene_ID'][i]
    if math.isnan(ncbi_id) == False :
        ncbi_id = int(ncbi_id)
        if symbol not in dic_ID_symbol2ncbi:
            dic_ID_symbol2ncbi[symbol] = ncbi_id
        else:
            print(symbol)
for i in range(0,ID_map.shape[0]):
    alias_list = str(ID_map["Alias_symbols"][i]).split(',')
    ncbi_id = ID_map['NCBI_Gene_ID'][i]
    ensemble_id = ID_map['Ensembl_gene_ID'][i]
    if len(alias_list) > 0:
        for alias in alias_list:
            if alias not in dic_ID_symbol2ncbi:
                dic_ID_symbol2ncbi[alias] = ncbi_id
dic_ID_symbol2ncbi['MEN1'] = 4221
dic_ID_symbol2ncbi['TSC1']= 7248
dic_ID_symbol2ncbi['WHSC1'] = 7468
dic_ID_symbol2ncbi['MET'] = 4233
dic_ID_symbol2ncbi['HIST1H1E'] = 3008
dic_ID_symbol2ncbi['FLNA'] = 2316
dic_ID_symbol2ncbi["FAM46D"] = 169966
dic_ID_symbol2ncbi["HIST1H1C"] = 3006
```

```python
df = pd.read_csv(input_dir1+"Tumor_specific_signatures.csv")
```

```python
df
```

```python
df_up = df.loc[df['SE'] > 0]
df_up = df_up.loc[df_up['median_rank'] > 0.75]
```

```python
result_df = pd.DataFrame()
result_df['object_name'] = df_up['Gene']
result_df['P_value'] = df_up['p_ttest']
```

```python
object_id_list = []
for name in list(df_up['Gene']):
    if name in dic_ID_symbol2ncbi:
        object_id_list.append(dic_ID_symbol2ncbi[name])
    else:
        object_id_list.append(None)
```

```python
result_df['object_id'] = object_id_list
```

```python
result_df['object_id_prefix'] = ['NCBIGene'] * len(object_id_list)
```

```python
result_df['object_category'] = ['Gene'] * len(object_id_list)
```

```python
result_df['subject_name'] = df_up['TumorType']
result_df['subject_id_prefix'] = ['MONDO']*len(object_id_list)
result_df['subject_category'] = ['Disease'] * len(object_id_list)


subject_id_list = []
for name in list(df_up['TumorType']):
    if 'TCGA_'+name in dic_mondo_map.keys():
        subject_id_list.append(dic_mondo_map['TCGA_'+name])
    else:
        subject_id_list.append(None)
```

```python
result_df['subject_id'] = subject_id_list
```

```python
result_df = result_df.dropna()
```

```python
result_df['object_id'] = result_df['object_id'].astype(int)
```

```python
result_df['predicate'] = ["has_biomarker"] * result_df.shape[0]
```

```python
result_df['Data_set'] = ["PMID: 27397505"]* result_df.shape[0]
```

```python
result_df['statistics_method'] = ['T_test'] * result_df.shape[0]
result_df
```

```python
result_df.to_csv("../Input_Table/GDSC_cancer_specific_signatures.csv", index = None)
```
